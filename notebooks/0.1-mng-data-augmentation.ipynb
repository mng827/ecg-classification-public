```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
sys.path.append(os.path.dirname(os.getcwd()))
from src.data.transforms import *
%load_ext autoreload
%autoreload 2
```

```python
train_df = pd.read_csv('../data/raw/mitbih_train.csv',header=None)
```

```python
label = ['Normal', 'Supraventricular', 'Ventricular', 'Fusion', 'Unknown']
num_samples = 5

fig, axes = plt.subplots(num_samples, figsize=(8,24))
train_examples = train_df.sample(n=num_samples)

for i in range(num_samples):
    signal = train_df.iloc[i, :-1].to_numpy()

    transform = GaussianNoise(magnitude_range=(0, 0.05), prob=1.0)
    signal_transformed = transform(signal)

    axes[i].plot(signal, label='Original')
    axes[i].plot(signal_transformed, label='Transformed')
    axes[i].set_title(label[int(train_df.iloc[i, -1])])
    axes[i].legend()
```
